```python
# assignment 
# Ceaser Chiper
# --------------------------------
# algorithm:
# get the input string 
# declare the shift value is 3 
# to encrypt the given text to chiper by getting the unicode of the char(ord()) and adding the shift value to it and converting it to char (chr())
```

```python
text = input("Enter the text to encrypt: ")
text_length = len(text)
shift = int(input("Enter the shift value: "))
encrypted_str =""

for char in text:
    shifted_char = chr((ord(char) + shift))
    encrypted_str += shifted_char
print (encrypted_str)
```
