```python
"""
Assignment
-----------
1) WAP to display the multiplication table from 10 to 1, for the first 10 tables
1 * 10 = 10
1 * 9  =  9
...

10 * 10 = 100
10 *  9 =  90
10 *  8 =  80
10 *  7 =  70
10 *  6 =  60
10 *  5 =  50
10 *  4 =  40
10 *  3 =  30
10 *  2 =  20
10 *  1 =  10

"""



first = 0 
while first < 10:
    first += 1 

    second = 10 
    while second >= 1:
        print(f"{first:2} * {second:2} = {first * second:3}",)
        second -= 1
        
    # print("=" * 20)  
```
